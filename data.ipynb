```python
from sqlalchemy.orm import sessionmaker
from model import Club, Training, Groupe, Championnat, engine

Session = sessionmaker(bind=engine)
session = Session()

results = session.query(Club, Training, Groupe, Championnat) \
    .join(Training, Club.id == Training.club_id) \
    .join(Groupe, Club.id == Groupe.club_id) \
    .join(Championnat, Groupe.championnat_id == Championnat.id) \
    .filter(Championnat.season == "2024-2025", Championnat.phase == "1") \
    .all()

data = []
for club, training, groupe, championnat in results:
    data.append({
        "club_name": club.club_name,
        "latitude": club.latitude,
        "longitude": club.longitude,
        "training_day": training.jour_entrainement,
        "training_start": training.debut_entrainement,
        "training_end": training.fin_entrainement,
        "group": groupe.groupe,
        "season": championnat.season,
        "phase": championnat.phase
    })

carto_df = pd.DataFrame(data)
carto_df.head()
```
